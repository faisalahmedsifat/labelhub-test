````python
import os
import re
# script to get all the functions from the files with .py extension and their associated comments and add them to the README.md file

# get all the files with .py extension
files = [f for f in os.listdir('.') if os.path.isfile(f) and f.endswith('.py')]
files
````

````python
import os

functions = []
for file in os.listdir('.'):
    if file.endswith('.py'):
        with open(file, 'r') as f:
            content = f.read()
            matches = re.findall(r'def\s+([^\(]+)\([^\)]*\)\s*:\s*("""[\s\S]*?""")', content)
            for match in matches:
                function_name = match[0]
                function_comment = match[1]
                functions.append((file, function_name, function_comment))

for (file, function_name, function_comment) in functions:
    print(file, function_name, function_comment)
    
    
# Path: README.md
"""
for every functions in functions:
put the file name, function name and function name in the following format in the README.md file:
--------------------------------------------------------------
## (bold) file name
### (bold) function name
function comment
--------------------------------------------------------------

"""
# Remove all the text from the README.md file
with open('README.md', 'w') as f:
    f.write('')
for (file, function_name, function_comment) in functions:
    print(file, function_name, function_comment)
    with open('README.md', 'a') as f:
        f.write('## `' + file + '`\n')
        f.write('### `' + function_name + '`\n')
        # remove """ from the function comment
        function_comment = function_comment[3:-3]
        f.write('```'+ function_comment + '\n')
        f.write('```\n')
        # f.write('--------------------------------------------------------------\n')
````
